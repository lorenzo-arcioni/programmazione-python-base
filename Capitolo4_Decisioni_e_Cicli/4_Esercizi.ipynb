## 📄 Esercizio 1: Classificatore di Numeri

### 📝 Traccia
Dato un numero intero, determina se è:
- Positivo, negativo o zero
- Pari o dispari (solo se non è zero)
- Piccolo (< 10), medio (10-100) o grande (> 100) in valore assoluto

Salva tutte le classificazioni in un dizionario, con chiavi: "segno", "parità", "grandezza", e stampalo.

````python
# Dati
numero = -42

# Inizializza il dizionario delle classificazioni
classificazioni = ...

##### Il tuo codice 🧑‍💻 #####
````

````python
# @title 👀 Soluzione Esercizio 1
numero = -42

# Inizializza il dizionario delle classificazioni
classificazioni = {}

# Determina segno
if numero > 0:
    classificazioni['segno'] = 'positivo'
elif numero < 0:
    classificazioni['segno'] = 'negativo'
else:
    classificazioni['segno'] = 'zero'

# Determina parità (solo se non è zero)
if numero != 0:
    if numero % 2 == 0:
        classificazioni['parità'] = 'pari'
    else:
        classificazioni['parità'] = 'dispari'

# Determina grandezza (valore assoluto)
valore_assoluto = abs(numero)
if valore_assoluto < 10:
    classificazioni['grandezza'] = 'piccolo'
elif valore_assoluto <= 100:
    classificazioni['grandezza'] = 'medio'
else:
    classificazioni['grandezza'] = 'grande'

print(f"Numero: {numero}")
print(f"Classificazioni: {classificazioni}")
````

## 📄 Esercizio 10: Countdown con Condizioni

### 📝 Traccia
Crea un countdown da un numero iniziale a 0, ma:
- Salta i numeri multipli di 3
- Se incontri un numero che finisce per 7, fermati immediatamente
- Conta quanti numeri hai effettivamente stampato

Ecco qui un esempio di output:

````python
############ INPUT ############

# numero_iniziale = 25

############ OUTPUT ############

# Countdown da 25:
# --------------------
# 25
# 23
# 22
# 20
# 19
# ⚠️ Fermato a 17 (finisce per 7)
# --------------------
# Numeri stampati: 5
````

````python
# Dati
numero_iniziale = 25
numeri_stampati = 0

##### Il tuo codice 🧑‍💻 #####

# Stampa informazioni finali

##### Il tuo codice 🧑‍💻 #####
````

````python
# @title 👀 Soluzione Esercizio 10

numero_iniziale = 25
numeri_stampati = 0

print(f"Countdown da {numero_iniziale}:")
print("-" * 20)

numero_corrente = numero_iniziale

while numero_corrente >= 0:
    # Controlla se finisce per 7
    if numero_corrente % 10 == 7:
        print(f"⚠️ Fermato a {numero_corrente} (finisce per 7)")
        break
    
    # Salta multipli di 3
    if numero_corrente % 3 == 0:
        numero_corrente -= 1
        continue
    
    # Stampa il numero
    print(numero_corrente)
    numeri_stampati += 1
    numero_corrente -= 1

# Stampa informazioni finali
print("-" * 20)
print(f"Numeri stampati: {numeri_stampati}")
````

````python
# @title 👀 Soluzione Esercizio 13
limite = 100
fibonacci = [0, 1]
numeri_pari = []
divisibili_per_3 = 0

print(f"Sequenza di Fibonacci fino a {limite}:")
print("-" * 40)

# Controlla i primi due numeri
for num in fibonacci:
    if num % 2 == 0:
        numeri_pari.append(num)
    if num % 3 == 0:
        divisibili_per_3 += 1

print(f"Inizio: {fibonacci}")

while True:
    # Calcola il prossimo numero di Fibonacci
    prossimo = fibonacci[-1] + fibonacci[-2]
    
    # Controlla se supera il limite
    if prossimo > limite:
        print(f"Prossimo numero ({prossimo}) supererebbe il limite")
        break
    
    # Aggiunge alla sequenza
    fibonacci.append(prossimo)
    print(f"Aggiunto: {prossimo}")
    
    # Controlla se è pari
    if prossimo % 2 == 0:
        numeri_pari.append(prossimo)
    
    # Controlla se è divisibile per 3
    if prossimo % 3 == 0:
        divisibili_per_3 += 1

# Stampa risultati
print("-" * 40)
print(f"Sequenza completa: {fibonacci}")
print(f"Numeri pari: {numeri_pari}")
print(f"Numeri divisibili per 3: {divisibili_per_3}")
print(f"Lunghezza sequenza: {len(fibonacci)}")
````

````python
# @title 👀 Soluzione Esercizio 14

numero = 195

print(f"Numero iniziale: {numero}")
passi = 0
numero_corrente = numero

while passi < 10:
    # Converti in stringa per controllare se è palindromo
    str_numero = str(numero_corrente)
    str_inverso = str_numero[::-1]
    
    if str_numero == str_inverso:
        print(f"Palindromo trovato: {numero_corrente} in {passi} passi!")
        break
    
    # Calcola l'inverso come numero
    numero_inverso = int(str_inverso)
    numero_precedente = numero_corrente
    numero_corrente = numero_corrente + numero_inverso
    passi += 1
    
    print(f"Passo {passi}: {numero_precedente} + {numero_inverso} = {numero_corrente}")

if passi == 10:
    print("Limite di 10 iterazioni raggiunto!")
````

````python
# Trova tutti i numeri primi fino a 100
primi = []  # Lista vuota dove salveremo tutti i numeri primi trovati
for n in range(2, 101):  # Itera tutti i numeri da 2 a 100 (incluso)

    è_primo = True  # Assume che n sia primo, fino a prova contraria
    for i in range(2, int(n**0.5) + 1):  # Controlla se n è divisibile per qualche numero da 2 alla radice quadrata di n
        if n % i == 0:  # Se n è divisibile, allora non è primo
            è_primo = False  # Imposta il flag a False
            break  # Interrompe il ciclo interno: non serve controllare oltre

    if è_primo:  # Se è rimasto True, allora n è primo
        primi.append(n)  # Aggiunge il numero primo alla lista
print(f"Primi trovati: {len(primi)}")  # Stampa quanti numeri primi sono stati trovati

# Trova le prime gemelle
gemelle = []  # Lista dove salveremo le coppie di primi gemelli

for i in range(len(primi) - 1):  # Scorriamo la lista dei numeri primi, fino al penultimo

    if primi[i + 1] - primi[i] == 2:  # Se la differenza tra due numeri primi consecutivi è 2
        gemelle.append((primi[i], primi[i + 1]))  # Allora sono "prime gemelle", le salviamo come tuple

print(f"\nPrime gemelle trovate: {len(gemelle)}")  # Stampa quante coppie di prime gemelle abbiamo trovato

for coppia in gemelle:  # Per ogni coppia trovata
    print(f"{coppia[0]}, {coppia[1]}")  # Stampa la coppia

# Coppia con somma più grande
if gemelle:  # Se abbiamo trovato almeno una coppia

    somma_max = 0  # Inizializza la somma massima trovata a 0
    coppia_max = None  # Inizializza la variabile per la coppia con somma massima

    for coppia in gemelle:  # Per ogni coppia di prime gemelle

        somma = coppia[0] + coppia[1]  # Calcola la loro somma
        if somma > somma_max:  # Se questa somma è maggiore della somma massima trovata finora

            somma_max = somma  # Aggiorna la somma massima
            coppia_max = coppia  # Salva la nuova coppia migliore

    print(f"\nCoppia con somma massima: {coppia_max} (somma: {somma_max})")  # Stampa la coppia con somma più alta

# Cerca triple gemelle
triple_trovate = False  # Flag per verificare se troviamo almeno una tripla gemella
for i in range(len(primi) - 2):  # Scorriamo la lista dei numeri primi, fino al terzultimo

    # Controlla se abbiamo tre numeri primi consecutivi con distanza 2
    if primi[i + 1] - primi[i] == 2 and primi[i + 2] - primi[i + 1] == 2:
        print(f"Tripla gemella trovata: {primi[i]}, {primi[i + 1]}, {primi[i + 2]}")  # Stampa la tripla
        triple_trovate = True  # Imposta il flag: abbiamo trovato almeno una tripla
        
if not triple_trovate:  # Se non abbiamo trovato nessuna tripla
    print("\nNessuna tripla gemella trovata fino a 100")  # Messaggio finale
````

````python
# Definisce una matrice 4x4 con valori interi
matrice = [
    [12, 7, 8, 14],
    [3, 22, 15, 6],
    [9, 4, 11, 2],
    [16, 13, 10, 18]
]

# Stampa l'intera matrice riga per riga
print("Matrice:")
for riga in matrice:
    print(riga)

# Inizializza una stringa vuota dove verrà costruito il messaggio
messaggio = ""

# --- Decodifica dal bordo della matrice in senso orario ---

# Visita la prima riga (indice 0), da sinistra verso destra
for j in range(4):
    valore = matrice[0][j]  # Estrae il valore dalla cella
    if valore % 2 == 0 and valore <= 52:  # Se è pari e non maggiore di 52
        lettera_num = valore // 2         # Divide per 2 per ottenere un numero da 1 a 26
        if 1 <= lettera_num <= 26:        # Se è nel range dell'alfabeto
            lettera = chr(ord('A') + lettera_num - 1)  # Converte in lettera (es. 1 → A)
            messaggio += lettera          # Aggiunge la lettera al messaggio

# Visita l'ultima colonna (indice 3), dall'alto verso il basso, escludendo l'angolo già letto
for i in range(1, 4):
    valore = matrice[i][3]
    if valore % 2 == 0 and valore <= 52:
        lettera_num = valore // 2
        if 1 <= lettera_num <= 26:
            lettera = chr(ord('A') + lettera_num - 1)
            messaggio += lettera

# Visita l'ultima riga (indice 3), da destra verso sinistra, escludendo l'angolo già letto
for j in range(2, -1, -1):
    valore = matrice[3][j]
    if valore % 2 == 0 and valore <= 52:
        lettera_num = valore // 2
        if 1 <= lettera_num <= 26:
            lettera = chr(ord('A') + lettera_num - 1)
            messaggio += lettera

# Visita la prima colonna (indice 0), dal basso verso l'alto, escludendo gli angoli già letti
for i in range(2, 0, -1):
    valore = matrice[i][0]
    if valore % 2 == 0 and valore <= 52:
        lettera_num = valore // 2
        if 1 <= lettera_num <= 26:
            lettera = chr(ord('A') + lettera_num - 1)
            messaggio += lettera

# Stampa il messaggio ottenuto dalla lettura del perimetro
print(f"\nMessaggio decodificato: {messaggio}")
````
