### ğŸ“„ Esercizio 11: Accumula fino a Soglia

#### ğŸ“ Traccia
Data una lista di numeri, usa un ciclo `while` per:
- Sommare i numeri uno alla volta
- Fermarsi quando la somma supera una soglia data
- Tenere traccia di quanti numeri sono stati sommati
- Stampa i risultati

Ecco un esempio di esecuzione:

````python
############ INPUT ############

# numeri = [5, 3, 8, 2, 12, 7, 4, 15, 6]
# soglia = 30

############ OUTPUT ############

# Lista: [5, 3, 8, 2, 12, 7, 4, 15, 6]
# Soglia: 30
# ------------------------------
# Aggiunto 5, somma parziale: 5
# Aggiunto 3, somma parziale: 8
# Aggiunto 8, somma parziale: 16
# Aggiunto 2, somma parziale: 18
# Aggiunto 12, somma parziale: 30
# Aggiungendo 7 supereremmo la soglia.
# ------------------------------
# Numeri utilizzati: [5, 3, 8, 2, 12]
# Somma finale: 30
# Numeri processati: 5
````

````python
# Dati
numeri = [5, 3, 8, 2, 12, 7, 4, 15, 6]
soglia = 30
somma = 0
indice = 0
numeri_utilizzati = []

##### Il tuo codice ğŸ§‘â€ğŸ’» #####

# Stampa risultati

##### Il tuo codice ğŸ§‘â€ğŸ’» #####
````

````python
#### ğŸ‘€ Soluzione Esercizio 11

numeri = [5, 3, 8, 2, 12, 7, 4, 15, 6]
soglia = 30
somma = 0
indice = 0
numeri_utilizzati = []

print(f"Lista: {numeri}")
print(f"Soglia: {soglia}")
print("-" * 30)

while indice < len(numeri) and somma <= soglia:
    numero_corrente = numeri[indice]
    
    # Controlla se aggiungendo questo numero supereremmo la soglia
    if somma + numero_corrente > soglia:
        print(f"Aggiungendo {numero_corrente} supereremmo la soglia.")
        break
    
    somma += numero_corrente
    numeri_utilizzati.append(numero_corrente)
    print(f"Aggiunto {numero_corrente}, somma parziale: {somma}")
    indice += 1

# Stampa risultati
print("-" * 30)
print(f"Numeri utilizzati: {numeri_utilizzati}")
print(f"Somma finale: {somma}")
print(f"Numeri processati: {len(numeri_utilizzati)}")
````

````python
#### ğŸ‘€ Soluzione Esercizio 12
parole = ["Casa", "gatto", "programmazione", "Python", "strada", "computer", "albero"]
indice = 0
parole_controllate = 0
parola_trovata = None

print("Ricerca della prima parola che:")
print("- Ha piÃ¹ di 5 caratteri")
print("- Contiene la lettera 'a'")
print("- Non inizia con maiuscola")
print("-" * 40)

while indice < len(parole):
    parola = parole[indice]
    parole_controllate += 1
    
    print(f"Controllo '{parola}':")
    
    # Controlla le condizioni
    ha_piu_5_caratteri = len(parola) > 5
    contiene_a = 'a' in parola.lower()
    non_inizia_maiuscola = not parola[0].isupper()
    
    print(f"  - PiÃ¹ di 5 caratteri: {ha_piu_5_caratteri}")
    print(f"  - Contiene 'a': {contiene_a}")
    print(f"  - Non inizia maiuscola: {non_inizia_maiuscola}")
    
    if ha_piu_5_caratteri and contiene_a and non_inizia_maiuscola:
        parola_trovata = parola
        print(f"  âœ… Trovata!")
        break
    else:
        print(f"  âŒ Non soddisfa tutti i criteri")
    
    print()
    indice += 1

# Stampa risultati
print("-" * 40)
if parola_trovata:
    print(f"Parola trovata: '{parola_trovata}'")
    print(f"Parole controllate: {parole_controllate}")
else:
    print("Nessuna parola trovata che soddisfi tutti i criteri")
    print(f"Parole controllate: {parole_controllate}")
````

### ğŸ“„ Esercizio 13: Sequenza di Fibonacci con While

#### ğŸ“ Traccia
Genera la sequenza di Fibonacci usando un ciclo `while` fino a quando non raggiungi un numero maggiore di 100. Inoltre: 

- Salva tutti i numeri pari della sequenza in una lista separata
- Conta quanti numeri della sequenza sono divisibili per 3
- Stampa la sequenza di Fibonacci, la sua lunghezza, i numeri pari e il numero di numeri divisibili per 3 nella sequenza.

Anche qui, esegui la cella *soluzione* per un esempio di esecuzione.

````python
# Dati
limite = 100
fibonacci = [0, 1]
numeri_pari = []
divisibili_per_3 = 0

##### Il tuo codice ğŸ§‘â€ğŸ’» #####

# Stampa risultati

##### Il tuo codice ğŸ§‘â€ğŸ’» #####
````

````python
#### ğŸ‘€ Soluzione Esercizio 13
limite = 100
fibonacci = [0, 1]
numeri_pari = []
divisibili_per_3 = 0

print(f"Sequenza di Fibonacci fino a {limite}:")
print("-" * 40)

# Controlla i primi due numeri
for num in fibonacci:
    if num % 2 == 0:
        numeri_pari.append(num)
    if num % 3 == 0:
        divisibili_per_3 += 1

print(f"Inizio: {fibonacci}")

while True:
    # Calcola il prossimo numero di Fibonacci
    prossimo = fibonacci[-1] + fibonacci[-2]
    
    # Controlla se supera il limite
    if prossimo > limite:
        print(f"Prossimo numero ({prossimo}) supererebbe il limite")
        break
    
    # Aggiunge alla sequenza
    fibonacci.append(prossimo)
    print(f"Aggiunto: {prossimo}")
    
    # Controlla se Ã¨ pari
    if prossimo % 2 == 0:
        numeri_pari.append(prossimo)
    
    # Controlla se Ã¨ divisibile per 3
    if prossimo % 3 == 0:
        divisibili_per_3 += 1

# Stampa risultati
print("-" * 40)
print(f"Sequenza completa: {fibonacci}")
print(f"Numeri pari: {numeri_pari}")
print(f"Numeri divisibili per 3: {divisibili_per_3}")
print(f"Lunghezza sequenza: {len(fibonacci)}")
````
